```python
from sklearn.preprocessing import MinMaxScaler

import numpy as np

import matplotlib.pyplot as plt

sample = 100

x = np.linspace(-1,5,sample)

x = x.reshape(sample,1)

z = (5.8)*(x)

noise = 0.3*np.random.randn(sample).reshape(100,1)

y = 1/(1+np.exp(-z))+noise

plt.scatter(x,y)

y = (y>=0.5)
y = y*1
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
```
