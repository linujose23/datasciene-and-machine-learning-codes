```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from matplotlib import style
from sklearn.datasets import make_blobs
style.use('ggplot')
%matplotlib inline
from scipy.spatial.distance import cdist          #for euclidean distance
import pandas as pd
maxx_iter = 400

K = 3
np.random.seed(0)
X = (make_blobs(n_samples=300,n_features=2,centers=K,cluster_std=2.1))[0]
centroids=  (make_blobs(n_samples=3,n_features=2,centers=K,cluster_std=2.1))[0]
plt.scatter(X[:,0],X[:,1])
plt.scatter(centroids[:,0],centroids[:,1])
```

```python
X[0]
```

```python
centroids = {}
```

```python
concat_centroids = np.concatenate((X[0],X[1],X[2]))
```

```python
centroids = np.array(concat_centroids.reshape(3,2))
```

```python
centroids
```

```python
distance = cdist(X,centroids,'euclidean') 
```

```python
min_distance = np.argmin(distance,axis=1)
```

```python
plt.scatter(X[:,0],X[:,1],c=min_distance)       #plotting our X and Y points
```

```python
one_hot_labels = np.array(pd.get_dummies(min_distance))            #one hot encoding min distances labels
```

```python
one_hot_labels.shape
```

```python
centroids.shape,X.shape
```

```python
for i in range(maxx_iter):

  #  distance = cdist(X,centroids,'euclidean') 

    e_d =(X - centroids.reshape(3,1,2))**2      
                         
    euclidean_distance = np.sqrt(np.sum(e_d,axis=2)).T      
    
    min_distance_allocation = np.argmin(euclidean_distance,axis=1)
    
    one_hot_sum = np.sum(one_hot_labels,axis=0)
    
    summ = np.dot(X.T,one_hot_labels)
    
    centroids = np.divide(summ,one_hot_sum).T
    
    
plt.scatter(X[:,0],X[:,1],marker='*',c=min_distance_allocation) 
plt.scatter(centroids[0][0],centroids[0][1],marker="o",s=150,color='r',cmap='rainbow')
plt.scatter(centroids[1][0],centroids[1][1],marker="o",s=150,color='b',cmap='rainbow')
plt.scatter(centroids[2][0],centroids[2][1],marker="o",s=150,color='g',cmap='rainbow')
```
